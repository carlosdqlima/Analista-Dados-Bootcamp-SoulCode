#### No colab faça a leitura desse arquivo em anexo e a partir dele crie 20 arquivos diferentes com os resultados de agrupamentos e salve esses arquivos como csv.

```python
import pandas as pd

df = pd.read_csv('/content/Base de dados - Dados sobre desgaste de funcionarios.csv')


# --------------------------------------------------------------------------------------------------------

'''
1 - Qual seria a taxa de rotatividade por departamento
'''

df['Demissão'] = df['Demissão'].map({'Sim': 1, 'Não': 0})

rotatividade_por_departamento = df.groupby('Departamento')['Demissão'].mean().reset_index()

rotatividade_por_departamento['Taxa de rotatividade departamento (%)'] = (rotatividade_por_departamento['Demissão'] * 100).round()

rotatividade_por_departamento = rotatividade_por_departamento.drop(columns=['Demissão'])

display(rotatividade_por_departamento)

rotatividade_dep = rotatividade_por_departamento.reset_index()
rotatividade_dep.to_csv('rotatividade_por_departamento.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
2 - Idade media por cargo
'''

idade_media_cargo = df.groupby('Cargo')['Idade'].mean().reset_index()
idade_media_cargo['Idade Média de Funcionarios'] = idade_media_cargo['Idade'].round()

idade_media_cargo = idade_media_cargo.drop(columns=['Idade'])
display(idade_media_cargo)

idade_media_cargo.to_csv('media_idade_cargo.csv', index= False)


# --------------------------------------------------------------------------------------------------------

'''
3 - Distância Média de Casa por Departamento
'''

distancia_media_dep = df.groupby('Departamento')['Distância de Casa'].mean().reset_index()
distancia_media_dep['Distância Média de Casa'] = distancia_media_dep['Distância de Casa'].round(3)
distancia_media_dep = distancia_media_dep.drop(columns=['Distância de Casa'])

display (distancia_media_dep)

distancia_media_dep.to_csv('distancia_media_casa_departamento.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
4 - Índice de Nível de envolvimento com o trabalho por Departamento
'''

indice_envolvimento_dep = df.groupby('Departamento')['Nível de envolvimento com o trabalho'].mean().reset_index()
indice_envolvimento_dep['Indice de Envolvimento'] = indice_envolvimento_dep['Nível de envolvimento com o trabalho'].round(2)
indice_envolvimento_dep = indice_envolvimento_dep.drop(columns='Nível de envolvimento com o trabalho')
display(indice_envolvimento_dep)

indice_envolvimento_dep.to_csv('indice_envolvimento_trabalho_departamento.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
5 - Proporção de Funcionários Ativos que Fazem Hora Extra por Departamento
'''

total_fazem_hora_extra_ativos = df[(df['Faz hora extra'] == 'Sim') & (df['Demissão'] == 0)].groupby('Departamento').size()

total_funcionarios_ativos = df[df['Demissão'] == 0].groupby('Departamento').size()

def calculo_proporcao(departamento):
  if departamento in total_fazem_hora_extra_ativos:
    return (total_fazem_hora_extra_ativos[departamento] / total_funcionarios_ativos[departamento]) * 100
  else:
    return 0

df['Proporção Hora Extra Departamento (%)'] = df['Departamento'].apply(calculo_proporcao).round()

proporcao_hora_extra_dep = (df[['Departamento', 'Proporção Hora Extra Departamento (%)']].drop_duplicates())
display(proporcao_hora_extra_dep)
proporcao_hora_extra_dep.to_csv('proporcao_funcionarios_fazem_hora_extra_por_departamentos.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
6 - Nível de Envolvimento com o Trabalho por Nível Hierárquico
'''

def classificar(nivel):
    if nivel <= 1:
      return 'Baixo'
    elif nivel >= 2 and nivel <= 3:
      return 'Médio'
    elif nivel >= 4:
      return 'alto'
    else:
        return 'Sem classificação'

df['Envolvimento com o Trabalho'] = df['Nível de envolvimento com o trabalho'].apply(classificar)

nivel_envolvimento_hieraquico = df.groupby('Nível hierárquico')['Envolvimento com o Trabalho'].agg(lambda x: x.mode()[0]).reset_index()

display(nivel_envolvimento_hieraquico)

nivel_envolvimento_hieraquico.to_csv('nivel_envolvimento_hieraquico.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
7 - Índice de Satisfação de Funcionarios com o Trabalho por Idade
'''

indice_satis_amb_trabalho = df.groupby('Idade')['Satisfação com o trabalho'].mean().round(2).reset_index()
display(indice_satis_amb_trabalho)

indice_satis_amb_trabalho.to_csv('indice_satisfacao_ambiente_idade.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
8 - Relação entre Escolaridade e Renda Mensal
'''

media_escolaridade_renda = df.groupby('Escolaridade')['Renda mensal'].mean().round(2).reset_index()

media_escolaridade_renda = media_escolaridade_renda.rename(columns={'Renda mensal': 'Renda Mensal Média'})
display(media_escolaridade_renda)
media_escolaridade_renda.to_csv('relacao_escolaridade_renda.csv', index= False)


# --------------------------------------------------------------------------------------------------------

'''
9 - Score de Performance por Departamento
'''

score_performance_dep = df.groupby('Departamento')['score de performance'].mean().round(2).reset_index()
score_performance_dep = score_performance_dep.rename(columns={'score de performance': 'Score de Performance Médio'})
display(score_performance_dep)
score_performance_dep.to_csv('score_performance_departamento.csv', index= False)

# --------------------------------------------------------------------------------------------------------

'''
10 - Quantidade Média de Treinamentos Realizados por Cargo
'''


qtd_media_treinamento_cargo = df.pivot_table(values='Qtde de treinamentos realizados no último ano', index='Cargo', aggfunc='mean').round(0).reset_index()
qtd_media_treinamento_cargo = qtd_media_treinamento_cargo.rename(columns={'Qtde de treinamentos realizados no último ano': 'Média de Qtde de Treinamentos Realizadas (LastYear)'})
display(qtd_media_treinamento_cargo)

qtd_media_treinamento_cargo.to_csv('media_qtde_treinamentos_cargos.csv', index = False)


# --------------------------------------------------------------------------------------------------------

'''
11 - Equilíbrio Vida-Trabalho por Gênero
'''

eq_vida_trabalho_genero = df.pivot_table(values='Equilíbrio vida trabalho',index='Gênero', aggfunc='mean').round(2).reset_index()
eq_vida_trabalho_genero = eq_vida_trabalho_genero.rename(columns={'Equilíbrio vida trabalho': 'Média de Equilíbrio Vida-Trabalho'})
display(eq_vida_trabalho_genero)
eq_vida_trabalho_genero.to_csv('equilibrio_vida_trabalho_genero.csv', index=False)


# --------------------------------------------------------------------------------------------------------

'''
12 - Anos Trabalhados na Empresa vs. Satisfação com o Trabalho
'''

anos_trabalhados_vs_satisf_trabalho = df.pivot_table(values='Satisfação com o trabalho', index='Total de anos trabalhados na empresa', aggfunc='mean').round(2).reset_index()
anos_trabalhados_vs_satisf_trabalho = anos_trabalhados_vs_satisf_trabalho.rename(columns = {'Satisfação com o trabalho':'Média Satisfação com o trabalho'})
display(anos_trabalhados_vs_satisf_trabalho)

anos_trabalhados_vs_satisf_trabalho.to_csv('media_satisfacao_anos_trabalho.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
13 - Tempo Médio desde a Última Promoção por Cargo
'''

tempo_medio_ultima_promocao_cargo = df.groupby('Cargo')['Anos desde a última promoção'].mean().round(1).reset_index()
tempo_medio_ultima_promocao_cargo = tempo_medio_ultima_promocao_cargo.rename(columns={'Anos desde a última promoção': 'Média de Anos Desde a Última Promoção'})
display(tempo_medio_ultima_promocao_cargo)
tempo_medio_ultima_promocao_cargo.to_csv('tempo_medio_ultima_promocao_cargo.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
14 - Anos Trabalhados com o Atual Gestor vs. Nível de Envolvimento
'''

anos_atual_gestor_envolvimento = df.pivot_table(values = 'Nível de envolvimento com o trabalho', index='Anos trabalhando com o atual gestor', aggfunc='mean').round(2).reset_index()
display (anos_atual_gestor_envolvimento)

anos_atual_gestor_envolvimento.to_csv('anos_atual_gestor_envolvimento.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
15 - Anos Trabalhados com o Atual Gestor vs. Nível de Satisfação com o ambiente de trabalho
'''

anos_atual_gestor_satisfacao_amb_trabalho = df.pivot_table(values = 'Nível de Satisfação com o ambiente de trabalho', index='Anos trabalhando com o atual gestor', aggfunc='mean').round(2).reset_index()
display (anos_atual_gestor_satisfacao_amb_trabalho)

anos_atual_gestor_satisfacao_amb_trabalho.to_csv('anos_atual_gestor_satisfacao_amb_trabalho.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
16 - Taxa de Demissões por Departamento
'''

taxa_demissoes_departamento = (df.groupby('Departamento')['Demissão'].mean()*100).reset_index(name='Taxa de Demissões (%)').round(2)

display(taxa_demissoes_departamento)

taxa_demissoes_departamento.to_csv('taxa_demissoes_departamento.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
17 - Satisfação com o Trabalho vs. Estado Civil
'''

satisfacao_trabalho_estado_civil = df.pivot_table(values='Satisfação com o trabalho', index='Estado civil', aggfunc='mean').reset_index().round(2)
display(satisfacao_trabalho_estado_civil)

satisfacao_trabalho_estado_civil.to_csv('satisfacao_trabalho_estado_civil.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
18 - Distribuição de Idade por Departamento
'''

distribuicao_idade_departamento = df.groupby('Departamento')['Idade'].mean().reset_index().round()
display(distribuicao_idade_departamento)

distribuicao_idade_departamento.to_csv('distribuicao_idade_departamento.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
19 - Comparação de Renda Mensal entre Funcionários de Diferentes Áreas de Formação
'''

comparacao_renda_areas_formacao = df.pivot_table(values = 'Renda mensal', index= 'Área de Formação', aggfunc='mean').reset_index().round(2)
display(comparacao_renda_areas_formacao)

comparacao_renda_areas_formacao.to_csv('comparacao_renda_areas_formacao.csv', index=False)

# --------------------------------------------------------------------------------------------------------

'''
20 - Diferentes medições de Salario por hora por Departamento
'''

medicoes_salario_hora_departamento = df.groupby('Departamento')['Salário por hora'].describe().round()
medicoes_salario_hora_departamento.columns = ['total', 'media', 'desvio_padrao', 'minimo', 'Q1', 'Mediana', 'Q3', 'maximo']

display(medicoes_salario_hora_departamento)

medicoes_salario_hora_departamento.to_csv('medicoes_salario_hora_departamento.csv')

# --------------------------------------------------------------------------------------------------------

```
