# Exercício 03 - Pandas


# Importação da biblioteca

```python
import pandas as pd
```

# Resolução Atividades

```python
# 1 - Crie um DataFrame com as colunas nome, idade, altura, peso e sexo usando listas.

nome = ['Carlos','Daniel','Júlia','Amanda','José','Alice']
idade = [21,20,22,19,24,42]
altura = [1.84,1.90,1.70,1.64,1.86,1.72]
peso = [74,70,65,70,90,72]
sexo = ['M','M','F','F','M','F']

df = pd.DataFrame(zip(nome,idade,altura,peso,sexo),
                  columns=['nome','idade','altura','peso','sexo'])
```

```python
# 2 - Exiba o DataFrame criado.

display(df)
```

```python
df.dtypes
```

```python
# 3 - Selecione e exiba a coluna "idade" do DataFrame.

df_idade = df.idade
print(df_idade)
```

```python
# 4 - Selecione e exiba a linha com índice 2 do DataFrame.
df_indice_2 = df.iloc[2]
display(df_indice_2)
```

```python
# 5 - Selecione e exiba a linha com rótulo 2 do DataFrame.

df_rotulo_2 = df.loc[2]
display(df_rotulo_2)
```

```python
# 6 - Selecione e exiba o valor da célula na linha de rótulo 3, coluna "altura".

df_linha_rotulo3_altura = df.loc[3,'altura']
display(df_linha_rotulo3_altura)
```

```python
# 7 - Renomeie as colunas "sexo" para "genero" e "peso" para "massa".

df.rename(columns={'sexo':'genero','peso':'massa'},inplace=True)
display(df)
```

```python
# 8 - Selecione e exiba as colunas "nome" e "altura" do DataFrame.

df_nome_altura = df[['nome','altura']]
display(df_nome_altura)
```

```python
# 9 - Filtre e exiba todas as linhas onde a idade é maior que 40.

df_idade_maior_40 = df[df['idade'] > 40]
display(df_idade_maior_40)
```

```python
# 10 - Adicione uma nova coluna chamada "IMC" ao DataFrame, calculando o Índice de Massa Corporal.

df['IMC'] = round(df['massa'] / df['altura'],2)
display(df)
```

```python
# 11 - Ordene o DataFrame pela coluna "altura" em ordem crescente.

df_ordenado_altura = df.sort_values('altura')
display(df_ordenado_altura)
```

```python
# 12 - Filtre e exiba todas as linhas onde a altura é  maior que 1.65.

df_altura_maior_1_65 = df[df['altura']>1.65]
display(df_altura_maior_1_65)
```

```python
# 13 - Modifique a altura de alguém para 1.91 e exiba o DataFrame.

df['altura'][0] = 1.91
display(df)
```

```python
# 14 - Crie uma coluna com valores de CPF para cada linha

import random

def gerar_cpf():
  return ''.join([str(random.randint(0, 9))
  for i in range(11)])

df['CPF'] = [gerar_cpf() for i in range(len(df))]

display(df)
```

```python
# 15 - Exclua a coluna CPF do DataFrame.

df = df.drop(columns=['CPF'])
```

```python
# 16 - Resete o índice do DataFrame e exiba o resultado.

df = df.reset_index(drop=True)
display(df)
```

```python
# 17 - Exiba as linhas do DataFrame onde a altura é maior que 1.70.

df_altura_maior_1_70 = df[df['altura']>1.70]
display(df_altura_maior_1_70)
```

```python
# 18 - Selecione e exiba as colunas "nome" e "genero".

df_nome_genero = df[['nome', 'genero']]
display(df_nome_genero)
```

```python
# 19 - Calcule e exiba a soma das idades.

soma_idades = df['idade'].sum()
display(soma_idades)
```

```python
# 20 - Salve o DataFrame em um arquivo CSV.

df.to_csv('exercicio_03_pandas.csv', index=False)
```
