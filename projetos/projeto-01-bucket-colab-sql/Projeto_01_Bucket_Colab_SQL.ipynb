```python
from typing import List

def definir_iam_publico_bucket(
    nome_bucket: str = "projeto-01-bucket-colab-sql",
    membros: List[str] = ["allUsers"],
):
    cliente_storage = storage.Client()
    bucket = cliente_storage.bucket(nome_bucket)

    politica = bucket.get_iam_policy(requested_policy_version=3)
    politica.bindings.append(
        {"role": "roles/storage.objectViewer", "members": membros}
    )

    bucket.set_iam_policy(politica)

print(f"O bucket {bucket.name} agora é publicamente legível")
```

```python
from google.cloud import storage
import pandas as pd
import io

# Inicialize o cliente do Cloud Storage
client = storage.Client()

# Defina o nome do bucket e o caminho do arquivo
bucket_name = 'projeto-01-bucket-colab-sql'
file_name = 'dados_brutos/balanco_petro.xlsx'  # Caminho completo do arquivo dentro do bucket

# Obtenha o bucket e o blob (arquivo)
bucket = client.bucket(bucket_name)
blob = bucket.blob(file_name)

# Faça o download do arquivo como bytes
data = blob.download_as_bytes()

# Carregue os dados em um DataFrame do pandas
# Corrected function name from read_exel to read_excel
df = pd.read_excel(io.BytesIO(data), skiprows=1)

# Exiba o DataFrame
print(df.head())
df.to_csv("balanco_petro.csv", index=False)
```

```python
df_tratado = df.T
df_tratado.columns = df.iloc[:, 0]
df_tratado = df_tratado.drop(df_tratado.index[0])
df_tratado.columns.name = None
df_tratado.index.name = None
df_tratado = df_tratado.reset_index()
df_tratado = df_tratado.rename(columns={'index': 'Datas Trimestrais'})
display(df_tratado)
```

```python
# Após a verificação de colunas duplicadas no balanço patrimonial, foi constatada algumas colunas com mesmo nome, acredito que essas colunas duplicadas tenha um propósito. No caso de "Contas a Receber" aparece tanto no Ativo Circulante quanto no Ativo Realizável a Longo Prazo, e isso é bem comum em demonstrações financeiras. Por essa razão não irei simplesmente eliminar as duplicatas, pois apresentam dados diferentes e podem ser relevantes, então irei realizar a renomeação dessas colunas para diferencia-las.

# Dicionário de mapeamento atualizado para renomear as colunas
dicionario_renomear = {
    'Contas a Receber': ['Contas a Receber (Circulante)', 'Contas a Receber (Longo Prazo)'],
    'Estoques': ['Estoques (Circulante)', 'Estoques (Longo Prazo)'],
    'Ativos Biológicos': ['Ativos Biológicos (Circulante)', 'Ativos Biológicos (Longo Prazo)'],
    'Despesas Antecipadas': ['Despesas Antecipadas (Circulante)', 'Despesas Antecipadas (Longo Prazo)'],
    'Empréstimos e Financiamentos': ['Empréstimos e Financiamentos (Circulante)', 'Empréstimos e Financiamentos (Não Circulante)'],
    'Passivos com Partes Relacionadas': ['Passivos com Partes Relacionadas (Circulante)', 'Passivos com Partes Relacionadas (Não Circulante)'],
    'Outros': ['Outros Passivos (Circulante)', 'Outros Passivos (Não Circulante)'],
    'Provisões': ['Provisões (Circulante)', 'Provisões (Não Circulante)'],
    'Passivos sobre Ativos Não-Correntes a Venda e Descontinuados': ['Passivos sobre Ativos Não-Correntes a Venda e Descontinuados (Circulante)', 'Passivos sobre Ativos Não-Correntes a Venda e Descontinuados (Não Circulante)'],
    'Tributos Diferidos': ['Tributos Diferidos (Ativo)', 'Tributos Diferidos (Passivo)'],
    'Adiantamento para Futuro Aumento Capital': ['Adiantamento para Futuro Aumento Capital (Passivo)', 'Adiantamento para Futuro Aumento Capital (PL)']
}

# Função para renomear as colunas com um contador
def renomear_colunas(col, contador):
    for nome_antigo, novo_nome in dicionario_renomear.items():
        if col == nome_antigo:
            novo_nome = novo_nome[contador[col]]
            contador[col] += 1
            return novo_nome
    return col

# Contador para rastrear quantas vezes cada coluna foi renomeada
contador = {key: 0 for key in dicionario_renomear.keys()}

# Aplicar a renomeação
df_tratado.columns = [renomear_colunas(col, contador) for col in df_tratado.columns]

print(df_tratado.columns)
```

```python
# renomeando as colunas para um formato mais amigavel para SQL

def renomear_colunas_sql(coluna):
  coluna = coluna.lower()
  coluna = coluna.replace(' ', '_')
  coluna = coluna.replace('(', '')
  coluna = coluna.replace(')', '')
  coluna = coluna.replace('/', '_')
  coluna = coluna.replace('-', '_')
  coluna = coluna.replace('á', 'a')
  coluna = coluna.replace('à', 'a')
  coluna = coluna.replace('é', 'e')
  coluna = coluna.replace('í', 'i')
  coluna = coluna.replace('ó', 'o')
  coluna = coluna.replace('ú', 'u')
  coluna = coluna.replace('ã', 'a')
  coluna = coluna.replace('õ', 'o')
  coluna = coluna.replace('ç', 'c')
  coluna = coluna.replace('â', 'a')
  coluna = coluna.replace('ê', 'e')
  coluna = coluna.replace('ô', 'o')
  return coluna

df_tratado.columns = [renomear_colunas_sql(col) for col in df_tratado.columns]

df_tratado.info()

df_tratado.to_csv("balanco_petro.csv", index=False)
```

```python
# Função para inserir dados do DataFrame na tabela
def inserir_dados(df_tratado, conexao):
    cursor = conexao.cursor()
    for _, row in df_tratado.iterrows():
        values = ', '.join(f"'{str(val)}'" for val in row)
        insert_query = f"INSERT INTO balanco ({', '.join(df_tratado.columns)}) VALUES ({values})"
        try:
            cursor.execute(insert_query)
        except Error as e:
            print(f"Erro ao inserir dados: {e}")
    conexao.commit()
    print("Dados inseridos com sucesso.")

inserir_dados(df_tratado, conexao)

# Função para consultar dados
def consultar_dados(conexao):
    try:
        query = "SELECT * FROM balanco"
        df_resultado = pd.read_sql(query, con=conexao)
        print(df_resultado)
    except Error as e:
        print(f"Erro ao consultar dados: {e}")

consultar_dados(conexao)

# Fechar a conexão
if conexao.is_connected():
    conexao.close()
    print("Conexão fechada.")
```
